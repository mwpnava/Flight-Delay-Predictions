```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/df_flight_weather_1.csv')
```

```python
#print head
df.head()
```

```python
#print null sum
print(df.isnull().sum())
```

```python
#print # of unique types for each column
val = {}
for col in df.columns:
    val[col] = len(df[col].unique())

for key, value in val.items():
    print(key, value)
```

```python
#for each column that has nulls and is data type string or object to 'NA'
for col in df.columns:
    if df[col].isnull().sum() > 0 and df[col].dtype == 'object':
        df[col].fillna('NA', inplace=True)
```
