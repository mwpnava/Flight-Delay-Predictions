```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load data 'datasets/kaggle/flights.csv'

df_f = pd.read_csv('../datasets/kaggle/flights.csv')
```

```python
#list all columns
df_f.columns
```

```python
#print size of the dataframe
# print(df_f.shape)
c_drop_non_numeric = ['DEPARTURE_TIME', 'SCHEDULED_DEPARTURE', 'ARRIVAL_TIME', 'SCHEDULED_ARRIVAL']
print(df_f.columns)

#convert all these columns to string, first remove everything after the decimal point if it exists, then convert to string, fill in missing placeholders with 0, string length should be 4
for col in c_drop_non_numeric:
    df_f[col] = df_f[col].astype(str).str.split('.').str[0].str.zfill(4)
    
#remove any row that has a character outside of the range 0-9
df_f = df_f[~df_f['DEPARTURE_TIME'].str.contains('[^0-9]')]
df_f = df_f[~df_f['SCHEDULED_DEPARTURE'].str.contains('[^0-9]')]
df_f = df_f[~df_f['ARRIVAL_TIME'].str.contains('[^0-9]')]
df_f = df_f[~df_f['SCHEDULED_ARRIVAL'].str.contains('[^0-9]')]
print(df_f.shape)
```

```python
# #if any value has '.', remove it and what is to the right of it
# df_f.loc[:, ['SCHEDULED_DEPARTURE', 'DEPARTURE_TIME', 'ARRIVAL_TIME', 'SCHEDULED_ARRIVAL']] = df_f.loc[:, ['SCHEDULED_DEPARTURE', 'DEPARTURE_TIME', 'ARRIVAL_TIME', 'SCHEDULED_ARRIVAL']].apply(lambda x: x.str.split('.').str[0])
# #for the entire df 'SCHEDULED_DEPATURE', 'DEPARTURE_TIME', 'ARRIVAL_TIME', 'SCHEDULED_ARRIVAL' add leading zeros to the values less than 4 digits
# df_f.loc[:, ['SCHEDULED_DEPARTURE', 'DEPARTURE_TIME', 'ARRIVAL_TIME', 'SCHEDULED_ARRIVAL']] = df_f.loc[:, ['SCHEDULED_DEPARTURE', 'DEPARTURE_TIME', 'ARRIVAL_TIME', 'SCHEDULED_ARRIVAL']].apply(lambda x: x.astype(str).str.zfill(4))

#temporarily convert to int, if any value is less than 0 or greater than 2359, remove it. Keep the column remove the row. convert back to string
df_f.loc[:, ['SCHEDULED_DEPARTURE', 'DEPARTURE_TIME', 'ARRIVAL_TIME', 'SCHEDULED_ARRIVAL']] = df_f.loc[:, ['SCHEDULED_DEPARTURE', 'DEPARTURE_TIME', 'ARRIVAL_TIME', 'SCHEDULED_ARRIVAL']].apply(lambda x: x.astype(int))
df_f = df_f[(df_f['SCHEDULED_DEPARTURE'] >= 0) & (df_f['SCHEDULED_DEPARTURE'] <= 2359)]
df_f = df_f[(df_f['DEPARTURE_TIME'] >= 0) & (df_f['DEPARTURE_TIME'] <= 2359)]
df_f = df_f[(df_f['ARRIVAL_TIME'] >= 0) & (df_f['ARRIVAL_TIME'] <= 2359)]
df_f = df_f[(df_f['SCHEDULED_ARRIVAL'] >= 0) & (df_f['SCHEDULED_ARRIVAL'] <= 2359)]
df_f.loc[:, ['SCHEDULED_DEPARTURE', 'DEPARTURE_TIME', 'ARRIVAL_TIME', 'SCHEDULED_ARRIVAL']] = df_f.loc[:, ['SCHEDULED_DEPARTURE', 'DEPARTURE_TIME', 'ARRIVAL_TIME', 'SCHEDULED_ARRIVAL']].apply(lambda x: x.astype(str))
#make sure to fill in the missing values with 0, 4 digits
df_f.loc[:, ['SCHEDULED_DEPARTURE', 'DEPARTURE_TIME', 'ARRIVAL_TIME', 'SCHEDULED_ARRIVAL']] = df_f.loc[:, ['SCHEDULED_DEPARTURE', 'DEPARTURE_TIME', 'ARRIVAL_TIME', 'SCHEDULED_ARRIVAL']].apply(lambda x: x.str.zfill(4))


#convert to 'HH:MM' format
df_f.loc[:, ['SCHEDULED_DEPARTURE', 'DEPARTURE_TIME', 'ARRIVAL_TIME', 'SCHEDULED_ARRIVAL']] = df_f.loc[:, ['SCHEDULED_DEPARTURE', 'DEPARTURE_TIME', 'ARRIVAL_TIME', 'SCHEDULED_ARRIVAL']].apply(lambda x: x.str[0:2] + ':' + x.str[2:4])
#remove any row that has a numeric value outside of the valid range HH:MM
```

```python
#turn 'YEAR' (2015), 'MONTH' (1-12). 'DAY' (1-31) into a single column 'DATE'
#turn 'DATE and 'SCHEDULED_DEPARTURE' (0-2359) into a single column 'SCHEDULED_DEPARTURE_TS'
#turn DATE and 'DEPARTURE_TIME' (0-2359) into a single column 'ACTUAL_DEPARTURE_TS'
#turn DATE and 'SCHEDULED_ARRIVAL' (0-2359) into a single column 'SCHEDULED_ARRIVAL_TS'
#turn DATE and 'ARRIVAL_TIME' (0-2359) into a single column 'ACTUAL_ARRIVAL_TS'

df_f['DATE'] = pd.to_datetime(df_f[['YEAR', 'MONTH', 'DAY']], format='%Y-%m-%d')
df_f['SCHEDULED_DEPARTURE_TS'] =  pd.to_datetime(df_f['DATE'].astype(str) + ' ' + df_f['SCHEDULED_DEPARTURE'].astype(str), format='%Y-%m-%d %H:%M')
df_f['ACTUAL_DEPARTURE_TS'] = pd.to_datetime(df_f['DATE'].astype(str) + ' ' + df_f['DEPARTURE_TIME'].astype(str), format='%Y-%m-%d %H:%M')
df_f['SCHEDULED_ARRIVAL_TS'] = pd.to_datetime(df_f['DATE'].astype(str) + ' ' + df_f['SCHEDULED_ARRIVAL'].astype(str), format='%Y-%m-%d %H:%M')
df_f['ACTUAL_ARRIVAL_TS'] = pd.to_datetime(df_f['DATE'].astype(str) + ' ' + df_f['ARRIVAL_TIME'].astype(str), format='%Y-%m-%d %H:%M')
```

```python
#remove columns where the 'ACTUAL_ARRIVAL_TS' is before the 'ACTUAL_DEPARTURE_TS' or the 'SCHEDULED_ARRIVAL_TS' is before the 'SCHEDULED_DEPARTURE_TS'
df_f = df_f[df_f['ACTUAL_ARRIVAL_TS'] >= df_f['ACTUAL_DEPARTURE_TS']]
df_f = df_f[df_f['SCHEDULED_ARRIVAL_TS'] >= df_f['SCHEDULED_DEPARTURE_TS']]
```

```python
#print out the first 5 rows of the dataframe of the new columns, due to a bug.
print(df_f[['DATE', 'SCHEDULED_DEPARTURE_TS', 'ACTUAL_DEPARTURE_TS', 'SCHEDULED_ARRIVAL_TS', 'ACTUAL_ARRIVAL_TS']].head())
```

```python
#list the count of non-null and null values for each column
# df_f.info()
def count_null(df):
    d = {}
    for col in df.columns:
        # 'col', 'count of non-null values', 'count of null values', 'percentage of null values'
    #    print(col, df_f[col].count(), df_f[col].isnull().sum(), df_f[col].isnull().sum()/df_f[col].count())
       d[col] = [df_f[col].count(), df_f[col].isnull().sum(), df_f[col].isnull().sum()/df_f[col].count()]

    return d

#call as a lambda function
d= count_null(df_f)
# print headers
print('col', 'count of non-null values', 'count of null values', 'percentage of null values')
for k,v in d.items():
    print(k, v[0], v[1], v[2])
```

```python
#remove column 'CANCELLATION REASON', appears 100% null
if 'CANCELLATION_REASON' in df_f.columns:
    df_f = df_f.drop('CANCELLATION_REASON', axis=1)
#at this point you can remove year
if 'YEAR' in df_f.columns:
    df_f = df_f.drop('YEAR', axis=1)

#remove 'flight number' and 'tail number' columns
if 'TAIL_NUMBER' in df_f.columns:
    df_f = df_f.drop('TAIL_NUMBER', axis=1)
if 'FLIGHT_NUMBER' in df_f.columns:
    df_f = df_f.drop('FLIGHT_NUMBER', axis=1)

#remove 'canceled' column
if 'CANCELLED' in df_f.columns:
    df_f = df_f.drop('CANCELLED', axis=1)

#remove 'diverted' column
if 'DIVERTED' in df_f.columns:
    df_f = df_f.drop('DIVERTED', axis=1)
```
