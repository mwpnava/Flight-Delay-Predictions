```python
import os
folder = '../datasets/wunderground/'
csv_files = [os.path.join(folder, file) for file in os.listdir(folder) if file.endswith('.csv')]
```

```python
import pandas as pd

#create a df, read all csv files, add a column for the file name without the extension or path
df = pd.concat((pd.read_csv(file).assign(filename = os.path.basename(file)) for file in csv_files), ignore_index = True)
```

```python
#describe
df.describe()
```

```python
#nulls
df.isnull().sum()
```

```python
#if these columns exist, drop them
#pressure_tend, pressure_desc, gust, max_temp, min_temp, precip_total, snow_hrly
if 'pressure_tend' in df.columns:
    df.drop(['pressure_tend'], axis = 1, inplace = True)

if 'pressure_desc' in df.columns:
    df.drop(['pressure_desc'], axis = 1, inplace = True)

if 'gust' in df.columns:
    df.drop(['gust'], axis = 1, inplace = True)

if 'max_temp' in df.columns:
    df.drop(['max_temp'], axis = 1, inplace = True)

if 'min_temp' in df.columns:
    df.drop(['min_temp'], axis = 1, inplace = True)

if 'precip_total' in df.columns:
    df.drop(['precip_total'], axis = 1, inplace = True)

if 'snow_hrly' in df.columns:
    df.drop(['snow_hrly'], axis = 1, inplace = True)
```

```python
# wdir
if 'wdir' in df.columns:
    #for null values in 'wdir' replace with 'NA'
    df['wdir'].fillna('?', inplace = True)

# wdir_cardinal
if 'wdir_cardinal' in df.columns:
    #for null values in 'wdir_cardinal' replace with 'NA'
    df['wdir_cardinal'].fillna('?', inplace = True)

# wx_phrase
if 'wx_phrase' in df.columns:
    #for null values in 'wx_phrase' replace with 'NA'
    df['wx_phrase'].fillna('?', inplace = True)


# day_ind 
if 'day_ind' in df.columns:
    #for null values in 'day_ind' replace with 'NA'
    df['day_ind'].fillna('?', inplace = True)


# uv_desc
if 'uv_desc' in df.columns:
    #for null values in 'uv_desc' replace with 'NA'
    df['uv_desc'].fillna('?', inplace = True)


# clds
if 'clds' in df.columns:
    #for null values in 'clds' replace with 'NA'
    df['clds'].fillna('?', inplace = True)
```

```python
# df.wx_phrase.value_counts( ascending=False)
#for each value print the count
def print_values(df, column):
    data  = {}
    for value in df[column].unique():
        data[value] = len(df[df[column] == value])

    return data

data = print_values(df, 'wx_phrase')
#sort the dictionary by value
sorted_data = sorted(data.items(), key=lambda x: x[1], reverse=True)
```

```python
# for key, value in sorted_data:
#     print(key, value)

#what percentage do the top 3 key values make up?
top_n = 15
print('Top {} values make up {}% of the data'.format(top_n, round(sum([value for key, value in sorted_data[:top_n]])/len(df)*100, 2)))

top_n_keys = [key for key, value in sorted_data[:top_n]]
```

```python
#for all other values, replace with 'other'
df['wx_phrase_simple'] = df['wx_phrase'].apply(lambda x: x if x in top_n_keys else 'other')
```

```python
data = print_values(df, 'wx_phrase')
#sort the dictionary by value
sorted_data = sorted(data.items(), key=lambda x: x[1], reverse=True)
for key, value in sorted_data:
    print(key, value)
```

```python
#how many rows have at least one null value?
print(f'{len(df[df.isnull().any(axis=1)])/len(df)*100}% of the data has at least one null value')

#remove rows with null values
# df.dropna(inplace = True)
#reset the index
df.reset_index(drop = True, inplace = True)
```

```python
#if expire_time_gmt is in the df, drop it
if 'expire_time_gmt' in df.columns:
    df.drop(['expire_time_gmt'], axis = 1, inplace = True)
if 'wx_phrase' in df.columns:
    df.drop(['wx_phrase'], axis=1, inplace=True)
```

```python
#convert 'valid_time_gmt' to day of the year 1-365
#convert to datetime '1.420120e+09' to '2015-01-01 00:00:00'
df['valid_time_gmt'] = pd.to_datetime(df['valid_time_gmt'], unit='s')
```

```python
#day of the year
df['day_of_year'] = df['valid_time_gmt'].dt.dayofyear

def doy_frequency(df):
    #add the frequency of each day of the year to a dictionary
    doy = {}
    for value in df['day_of_year'].unique():
        doy[value] = len(df[df['day_of_year'] == value])
    return doy

#print the frequency of each day of the year
doy = doy_frequency(df)
for key, value in doy.items():
    print(key, value)
```
