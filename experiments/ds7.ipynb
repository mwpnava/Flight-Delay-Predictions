```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/df_flight_weather_6.csv')
df.head()
```

```python
#delete the columns that are not needed
#WHEELS_OFF, WHEELS_ON, TAXI_IN, TAXI_OUT
if 'WHEELS_OFF' in df.columns:
    df.drop(['WHEELS_OFF'], axis=1, inplace=True)
if 'WHEELS_ON' in df.columns:
    df.drop(['WHEELS_ON'], axis=1, inplace=True)
if 'TAXI_IN' in df.columns:
    df.drop(['TAXI_IN'], axis=1, inplace=True)
if 'TAXI_OUT' in df.columns:
    df.drop(['TAXI_OUT'], axis=1, inplace=True)
if 'DISTANCE' in df.columns:
    df.drop(['DISTANCE'], axis=1, inplace=True)
```

```python
#describe df
vals = {}
for col in df.columns:
    vals[col] = [df[col].isna().sum(), df[col].nunique(), df[col].dtype]
print("Name", "Null", "Unique", "Type")
for k, v in vals.items():
    print(k, v)
```

```python
df['month'] = df['month'].astype('category')
df['day_of_month'] = df['day_of_month'].astype('category')
df['day_of_week'] = df['day_of_week'].astype('category')
df['day_of_year'] = df['day_of_year'].astype('category')
```

```python
#save as csv
df.to_csv('../datasets/df_flight_weather_7.csv', index=False)
```
