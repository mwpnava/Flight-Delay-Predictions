```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/df_flight_weather_7.csv')
df.head()
```

```python
#print columns
df.columns
# df.head()
```

```python
#remove columns
#ELAPSED_TIME, AIR_TIME
if 'ELAPSED_TIME' in df.columns:
    df.drop(['ELAPSED_TIME'], axis=1, inplace=True)
if 'AIR_TIME' in df.columns:
    df.drop(['AIR_TIME'], axis=1, inplace=True)

#DEPARTURE_DELAY
if 'DEPARTURE_DELAY' in df.columns:
    df.drop(['DEPARTURE_DELAY'], axis=1, inplace=True)

    #WX_ICON_DEST, WX_ICON
if 'WX_ICON_DEST' in df.columns:
    df.drop(['WX_ICON_DEST'], axis=1, inplace=True)

if 'WX_ICON' in df.columns:
    df.drop(['WX_ICON'], axis=1, inplace=True)
    
```

```python
#save csv
df.to_csv('../datasets/df_flight_weather_8.csv', index=False)
```
