```python
df_2 =pd.read_csv('../datasets/kaggle/flights.csv')
df_2.head()
```

```python
#list all unique numeric values in the column
values = df_2['ORIGIN_AIRPORT'].unique()
#print leng of df_2
print(len(df_2))
#if value is numeric, count
count = 0
numeric_values = []
for value in values:
    if isinstance(value, (int, float)):
        count += 1
        numeric_values.append(value)
#print length of unique values
print('{} out of {} values are numeric'.format(count, len(values)))
#remove all numeric values from the list
for value in numeric_values:
    values = values[values != value]
#only keep values that are not numeric
df_2 = df_2[df_2['ORIGIN_AIRPORT'].isin(values)]
#print length of df_2
print(len(df_2))

```
